```python
cliente = pd.read_excel('cliente_varejo.xlsx')
cliente.head()
```

```python
venda_correto.head()
```

```python
cliente[['idade', 'renda']].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({'renda': 'float'})
```

```python
cliente.info()
```

```python
vendas_cliente = vendas_correto.merge(cliente, how='left', on='cliente_Log')
vendas_cliente.head()
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```python
agg_idade_bandeira = round(vendas_cliente.groupby('bandeira')['idade'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(4, 2))

plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'], color='blue')

plt.ylabel('Média de Idade')
plt.xlabel('Bandeira')

plt.title('Media de idade por Bandeira', loc='center', fontsize=18, fontweight='bold', color='black');
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_idcanal_renda['idcanalvenda'], agg_idcanal_renda['renda'], color='yellow')
plt.ylabel('Média de renda')
plt.xlabel('Canal')
plt.title('Media de renda por canal', loc='center', fontsize=10, fontweight='bold', color='black');
```

```python
venda_data = vendas_cliente.groupby('Data').idcompra.nunique().reset_index()
venda_data
```

```python
fig = plt.figure(figsize=(15,9))

plt.ylabel('Data da venda')
plt.xlabel('Qtd de vendas')

y = (venda_data['idcompra'])
x = (venda_data['Data'])

plt.plot(x, y, color = 'green', linewidth = '2');
```

```python
departamento_preco = round(vendas_cliente.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
departamento_preco
```

```python
fig = plt.figure(figsize=(15,4))
plt.bar(departamento_preco['Nome_Departamento'], departamento_preco['Preço_com_frete'], color='green')
plt.ylabel('Média Preço com Frete')
plt.xlabel('Departamento')
plt.title('Media do Preço com frete por departamento', loc='center', fontsize=10, fontweight='bold', color='black');
```
